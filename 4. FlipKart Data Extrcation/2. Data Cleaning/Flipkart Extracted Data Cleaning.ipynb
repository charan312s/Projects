```python
# Data Cleaning

# Drop unnecessary columns
cleaned_data = flipkart_data.drop(columns=['Unnamed: 0', 'image_url'])

# Clean the price columns (remove currency symbols and commas, convert to numeric)
cleaned_data['actual_price'] = cleaned_data['actual_price'].replace('[₹,]', '', regex=True).astype(int)
cleaned_data['selling_price'] = cleaned_data['selling_price'].replace('[₹,]', '', regex=True).astype(int)

# Extract numeric values from 'no_of_ratings' and 'no_of_reviews'
cleaned_data['no_of_ratings'] = cleaned_data['no_of_ratings'].str.extract('(\d+[,]*\d*)').replace(',', '', regex=True).astype(int)
cleaned_data['no_of_reviews'] = cleaned_data['no_of_reviews'].str.extract('(\d+[,]*\d*)').replace(',', '', regex=True).astype(int)


# Display the cleaned data
cleaned_data.head()
```

```python
# Function to extract memory information and clean the mobile name
def extract_memory_and_name(row):
    # Extract memory (e.g., "128 GB", "12 GB RAM | 256 GB ROM") using regex
    memory_match = re.search(r'(\d+ GB(?: RAM)?(?: \| \d+ GB ROM)?)', row['name'])
    memory = memory_match.group(0) if memory_match else 'Unknown'

    # Extract the mobile mobile name (remove memory and any trailing commas)
    model_name = re.sub(r'\s*\(.*\)\s*$', '', row['name']).strip()

    return pd.Series([model_name, memory])

# Apply the function to the dataset
cleaned_data[['Mobile_name', 'extracted_memory']] = cleaned_data.apply(extract_memory_and_name, axis=1)

# Drop the original 'name' column
cleaned_data = cleaned_data.drop(columns=['name'])

# Display the cleaned dataset
cleaned_data.head()
```

```python
import pandas as pd
import re

# Function to extract RAM, ROM, and expandable memory
def extract_memory_features(memory):
    # Ensure memory is a string
    memory = str(memory)
    
    # Extract RAM
    ram_match = re.search(r'(\d+\s*GB\s*RAM)', memory)
    ram = ram_match.group(0).replace(' RAM', '') if ram_match else 'Unknown'
    
    # Extract ROM
    rom_match = re.search(r'(\d+\s*GB\s*ROM)', memory)
    rom = rom_match.group(0).replace(' ROM', '') if rom_match else 'Unknown'
    
    # Extract expandable memory
    expandable_match = re.search(r'Expandable\s*Upto\s*(\d+\s*GB)', memory)
    expandable_memory = expandable_match.group(1) if expandable_match else 'Not Expandable'
    
    return pd.Series([ram, rom, expandable_memory])

# Convert memory column to string
cleaned_data['memory'] = cleaned_data['memory'].astype(str)

# Apply the function to extract RAM, ROM, and expandable memory
cleaned_data[['RAM(in GB)', 'ROM(in GB)', 'expandable_memory']] = cleaned_data['memory'].apply(extract_memory_features)

# Drop the original 'memory' column
cleaned_data = cleaned_data.drop(columns=['memory'])

# Display the cleaned dataset
cleaned_data.head()
```

```python
import re

# Function to extract screen size and specific display type
def extract_screen_features(dimension):
    # Ensure dimension is a string
    dimension = str(dimension)
    # Extract screen size (e.g., "15.49 cm")
    size_match = re.search(r'(\d+(\.\d+)?\s*cm)', dimension)
    screen_size = size_match.group(0) if size_match else 'Unknown'

    # Extract display type by searching for common display types
    display_type_patterns = [
        r'AMOLED', r'LCD', r'Retina', r'Super Retina', r'OLED', r'IPS', r'TFT',
        r'Dynamic AMOLED', r'Super AMOLED', r'Liquid Retina', r'P-OLED', r'Full HD', r'HD'
    ]
    display_type = 'Normal Display'  # Default value
    for pattern in display_type_patterns:
        if re.search(pattern, dimension, re.IGNORECASE):
            display_type = re.search(pattern, dimension, re.IGNORECASE).group(0)
            break

    return pd.Series([screen_size, display_type])

# Apply the function to extract screen size and display type
cleaned_data[['screen_size', 'display_type']] = cleaned_data['dimension'].apply(extract_screen_features)

# Drop the original 'dimension' column
cleaned_data = cleaned_data.drop(columns=['dimension'])

# Display the cleaned dataset
cleaned_data.head()
```

```python
import pandas as pd
import re

def process_camera(camera_info):
    # Ensure camera_info is a string
    camera_info = str(camera_info)
    # Split the camera information into rear and front parts
    parts = camera_info.split('|')
    rear_camera_part = parts[0].strip()
    front_camera_part = parts[1].strip() if len(parts) > 1 else "Unknown"
    
    # Extract rear camera MP values and sum them up
    rear_camera_values = [float(match[0]) for match in re.findall(r'(\d+(\.\d+)?)MP', rear_camera_part)]
    rear_camera_sum = sum(rear_camera_values) if rear_camera_values else "Unknown"
    
    # Extract front camera MP value
    front_camera_value = re.search(r'(\d+(\.\d+)?)MP', front_camera_part)
    front_camera = front_camera_value.group(0) if front_camera_value else "Unknown"
    
    return pd.Series([f"{rear_camera_sum}MP", front_camera])

# Apply the function to process the camera column
cleaned_data[['rear_camera(in MP)', 'front_camera(in MP)']] = cleaned_data['camera'].apply(process_camera)

# Drop the original 'camera' column if no longer needed
cleaned_data = cleaned_data.drop(columns=['camera'])

# Display the processed data
cleaned_data.head()
```

```python
import pandas as pd
import re

# Assuming cleaned_data is already defined
def extract_battery_capacity(battery_str):
    # Ensure battery_str is a string
    battery_str = str(battery_str)
    # Use regex to find integers in the battery string
    match = re.search(r'(\d+)', battery_str)
    if match:
        return int(match.group(1))
    else:
        return None  # or return a default value if needed

# Apply the function to the battery column
cleaned_data['battery'] = cleaned_data['battery'].apply(extract_battery_capacity)

# Display the updated DataFrame
cleaned_data.head()
```

```python
import pandas as pd
import re

# Function to convert warranty to years
def convert_warranty_to_years(warranty_str):
    # Ensure warranty_str is a string
    warranty_str = str(warranty_str)
    # Extract numeric values (e.g., 1 Year, 6 Months)
    match = re.search(r'(\d+)\s*(Year|Month)', warranty_str)
    if match:
        value = int(match.group(1))
        unit = match.group(2)
        if unit == 'Month':
            return value / 12  # Convert months to years
        else:
            return value  # Already in years
    return None  # Return None if no match found

# Apply the function to the warranty column
cleaned_data['warranty'] = cleaned_data['warranty'].apply(convert_warranty_to_years)

# Display the updated DataFrame
cleaned_data.head()
```

```python
def extract_company_name(mobile_name):
    # Split the model name by spaces and return the first part
    return mobile_name.split()[0]

# Apply the function to the dataset
cleaned_data['Model'] = cleaned_data['Mobile_name'].apply(extract_company_name)

# Display the cleaned dataset with company names
cleaned_data.head()
```

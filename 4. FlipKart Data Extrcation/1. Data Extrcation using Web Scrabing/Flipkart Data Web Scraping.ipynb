```python


import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"

def getdata(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text)


    '''main_division = soup.find_all(class_="_1YokD2 _3Mn1Gg")
    main = main_division[0]'''
    # Lets extract the class that is related to the name of the mobile
    mobile_names = soup.find_all(class_="_4rR01T")
    # 1. Name of the mobile list
    name = []
    for x in mobile_names:
        name.append(x.text)
    


    # Lets extract the class that is related to the rating out of 5, number of reviews and rating
    review = soup.find_all(class_="_2_R_DZ")
    # 2. Number of Rating list
    ratings = []
    # 3. Number of Reviews list
    reviews = []

    for row in review:
        if len(ratings) <=24:            # or if i<=24 #Since there are other ratings and reviews are available in the page with same class
            temp1 = row.span.span
            temp2 = temp1.next_sibling.next_sibling
            ratings.append(temp1.text)
            reviews.append(temp2.text)   # i++
        else:
            break



    # Now let's extract the rating out of 5 for each product
    # 4. Rating Out of 5 list 
    rating_outof_5 = []

    for row in review:
        previous = row.previous_sibling
        temp = previous.div
        rating_outof_5.append(temp.text)
    

    '''actual_price = [None for i in range(24)]
    selling_price = [None for i in range(24)]
    offer = [None for i in range(24)]'''
    
    
    '''# Lets extract the class that is related to the price and offer details
    price = soup.find_all(class_="_30jeq3 _1_WHN1")
    # 5. Currently, Selling price list
    selling_price = []
    for tag in price:
        selling_price.append(tag.text)
    
    # 6. Currently, Actual Price list
    actual_price = []
    # 7. Currently, Mobile Offer list
    offer = []

    for row in price:
        print(row)
        temp1 = row.next_sibling
        print(temp1)
        actual_price.append(temp1.get_text())
        temp2 = temp1.next_sibling
        print(temp2)
        temp3 = temp2.find('span')
        #print(temp3)
        offer.append(temp3.text)'''
    selling_price = []
    actual_price = []
    offer = []
    
    a = soup.find_all(class_='_30jeq3 _1_WHN1')
    b = soup.find_all(class_='_3I9_wc _27UcVY')
    c = soup.find_all(class_='_3Ay6Sb')
    
    '''print(a[:5])
    print(b[:5])
    print(c[:5])'''
    for i in a:
        selling_price.append(i.text)
    for i in b:
        actual_price.append(i.text)
        offer.append(i.next_sibling.text)
    '''for i in c:
        offer.append(i.text)'''

    # Lets extract the class that is related to the image url of the image
    image = soup.find_all(class_="CXW8mj")
    # 8. Mobile image url list
    image_url = []

    i = 1
    for row in image:
        if  i<=24:
            tag = row.img
            link = tag.src
            image_url.append(link)
            i += 1
        else:
            break



    # Lets extract the class that is related to the specifications of the mobile
    specifications = soup.find_all(class_="_1xgFaf")
    
    # The specifications variable consists a list of specifications for each mobile
    # The important observation in specifications of each mobile is that - 
    # Some of the mobiles consists only 5 specifications and others consists of 6
    # Inorder to avoid the confusion about the specifications of each individual mobile
    # Lets, Create a dictionary which stores the specifications of individual mobile phone
    # The keys of the dictionary are the range of numbers from 1 to 25(total=24)
    # The values of individual keys are the list of specifications of individual mobile phones
    # This  dictionary helps to avoid the confusion and to extract and store the data in the respective variables in an easiest way

    dic = {}
    for i in range(1,25):
        dic[i] = []

    i = 1
    for row in specifications:
        temp = row.find_all(class_="rgWa7D")
        for tag in temp:
            text = tag.text
            dic[i].append(text)
        i += 1

    # These are the specifications that each product is having in the flipkart site
    # 9. Memory Capacity of the mobile
    memory = []
    # 10. Dimension and display information of the mobile
    length = []
    # 11. Camera Specification of the mobile
    camera = []
    # 12. Battery Information
    battery = []
    # 13. Processor Specification
    processor = []
    # 14. Warrenty Information
    warranty = []

    for key in dic.keys():
        if len(dic[key]) == 6:
            memory.append(dic[key][0])
            length.append(dic[key][1])
            camera.append(dic[key][2])
            battery.append(dic[key][3])
            processor.append(dic[key][4])
            warranty.append(dic[key][5])
        elif len(dic[key]) == 5:
            memory.append(dic[key][0])
            length.append(dic[key][1])
            camera.append(dic[key][2])
            battery.append(None)
            processor.append(dic[key][3])
            warranty.append(dic[key][4])
        else:
            memory.append(None)
            length.append(None)
            camera.append(None)
            battery.append(None)
            processor.append(None)
            warranty.append(None)




    # Lets extract the class that is related to the product url
    link = soup.find_all(class_="_1fQZEK")
    # 15. Product url for more information
    product_link = []

    for row in link:
        temp = row.get('href')
        url = "https://www.flipkart.com"+temp
        product_link.append(url)



    # Alright, Now we are having a set of 15 attributes
    # It's the to use all of tose and make those into a datafram
    # Let's do it!!!
    dictionary = {"name": name, "actual_price": actual_price, "selling_price": selling_price, "offer": offer,
                        "rating(outof5)": rating_outof_5, "no_of_ratings": ratings, "no_of_reviews": reviews, "image_url": image_url,
                        "memory": memory, "dimension": length, "camera": camera, "battery": battery, "processor": processor,
                        "warranty": warranty, "product_link": product_link}

    for key in dictionary.keys():
        print(key)
        print(len(dictionary[key]))
    print('\n')
    dataframe = pd.DataFrame(dictionary)


    # Work was done!! It's time returnn the dataframe

    return dataframe



data = getdata(url)

data.head()


```
